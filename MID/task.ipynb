```python
import tensorflow
from tensorflow.keras.datasets import cifar10
import matplotlib.pyplot as plt
from tqdm import tqdm
import os
import cv2
from google.colab.patches import cv2_imshow

```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python

# Display the image
plt.imshow(x_test[0])
plt.title(f"Label: {y_test[0]}")
plt.show()
```

```python
index = 0  # Change this to the index of the image you want to display

# Display the image
resize_img=cv2.resize(x_test[index],(400,400))
cv2_imshow(resize_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

# Convert all images to grayscale
grayscale_images = []
for index in tqdm(range(len(x_test))):
    grayscale_image = cv2.cvtColor(x_test[index], cv2.COLOR_BGR2GRAY)
    grayscale_images.append(grayscale_image)



```

```python
index = 0  # Change this to the index of the image you want to display

# Display the image
resize_img=cv2.resize(grayscale_images[index],(400,400))
cv2_imshow(resize_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
